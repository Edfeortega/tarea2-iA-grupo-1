```python
#Tarea 2
#Inteligencia Artificial 2024-1, Grupo 1
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1. Loading your friend's data into a dictionary
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
durations = [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]

movie_dict = {"years": years, "durations": durations}

movie_dict
```

```python
#2. Creating a DataFrame from a dictionary
durations_df = pd.DataFrame(movie_dict)

durations_df
```

```python
#3. A visual inspection of our data
datos = {
    'Año': [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    'Duración_promedio': [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]
}
df_netflix = pd.DataFrame(datos)


fig, ax = plt.subplots()


ax.plot(df_netflix['Año'], df_netflix['Duración_promedio'], marker='o', linestyle='-')


ax.set_title('Tendencia de la Duración Promedio de Películas en Netflix (2011-2020)')
ax.set_xlabel('Año')
ax.set_ylabel('Duración Promedio (minutos)')
plt.grid()

plt.show()
```

```python
#4. Loading the rest of the data from a CSV
# Lee el archivo CSV como un DataFrame
netflix_df = pd.read_csv('netflix_data.csv')

# Imprime las primeras cinco filas del DataFrame
netflix_df.head(5)
```

```python
#5. Filtering for movies!
netflix_df_movies_only = netflix_df[netflix_df['type'] == 'Movie']

netflix_movies_col_subset = netflix_df_movies_only[['title', 'country', 'genre', 'release_year', 'duration']]


netflix_movies_col_subset.head()
```

```python
#6. Creating a scatter plot
netflix_df = pd.read_csv('netflix_data.csv')

fig = plt.figure(figsize=(12, 8))


plt.scatter(netflix_df['release_year'], netflix_df['duration'], alpha=0.5)


plt.title('Duración de las Películas de Netflix en Función del Año de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Duración (minutos)')
plt.grid()

plt.show()
```

```python
#7. Digging deeper
duration_less_60min = netflix_movies_col_subset[netflix_movies_col_subset['duration'] < 60]


duration_less_60min.head(20)
```

```python
#8. Marking non-feature films
colors = []


for genre in netflix_movies_col_subset['genre']:
    if genre in ['Children', 'Stand-Up', 'Documentaries']:
        colors.append('blue')  
    else:
        colors.append('black')  


colors[:10]
```

```python
#9. Plotting with color!

fig = plt.figure(figsize=(12, 8))
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'], c=colors, alpha=0.5)
plt.title('Duración de las Películas de Netflix en Función del Año de Lanzamiento (Con Colores por Género)')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Duración (minutos)')
plt.grid()
plt.show()
```
